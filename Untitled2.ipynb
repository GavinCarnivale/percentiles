```python
import numpy as np
import matplotlib.pyplot as plt

# Load data: skip header and first row of data (assuming you want height column, index 1)
data = np.loadtxt("SOCR-HeightWeight.csv", delimiter=",", dtype=float, skiprows=1)
data = data[1:, 1].astype(float)  # Adjust if you want a different column or if skipping is different

# Calculate quartiles
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)
Q3 = np.percentile(data, 75)

# Calculate IQR
IQR = Q3 - Q1

# Calculate bo
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data < lower_bound) | (data > upper_bound)]

plt.figure(figsize=(7, 7))
plt.boxplot(data, vert=False)
plt.title("Boxplot")
plt.show()

print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
print("Min value:", np.min(data))
print("Max value:", np.max(data))
print(f"Lower bound for outliers: {lower_bound}")
print(f"Upper bound for outliers: {upper_bound}")
print("Number of outliers:", len(outliers))

```
